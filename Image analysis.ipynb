### Overview

We have selected the <a href="https://github.com/zalandoresearch/fashion-mnist">Fashion MNIST dataset</a>. This dataset consists of 70,000 labeled fashion images. The data is divided into 10 catagories. As per the git hub page this dataset was created to replace the MNIST dataset which consists of handwritten images.

The creators of the Fashion-MNIST created this dataset to serve as direct drop-in replacement for the MNIST dataset and as such this dataset has same size and structure.

The authors of the dataset sited 3 reasons why they thought of relacing the MNIST dataset.
1. MNIST is too easy. Convolutional nets can achieve 99.7% on MNIST. Classic machine learning algorithms can also achieve 97% easily.
2. MNIST is overused. In <a href="https://twitter.com/goodfellow_ian/status/852591106655043584">this April 2017 Twitter thread</a>, Google Brain research scientist and deep learning expert Ian Goodfellow calls for people to move away from MNIST.
3. MNIST can not represent modern Computer Vision tasks, as noted in <a href="https://twitter.com/fchollet/status/852594987527045120">this April 2017 Twitter thread</a>, deep learning expert/Keras author Fran√ßois Chollet.

### Prediction accuracy

The purpose of the dataset is to benchmark the performance of the machine learning algorithms, hence  we cannot say that a perticular accuracy of prediction should be achieved by the algorithm to be considered usefull. In practice most of the CV related machine learning algorithms are initally tested on the MNIST dataset to check if the algorithm will actually work. The kaggel discription of the fashion-MNIST dataset has a quote about the MNIST dataset: "If it doesn't work on MNIST, it won't work at all" and it further goes on to say that successfull teston MNIST does not garuntee that same algorithm will not fail on another dataset, exact quote being "Well, if it does work on MNIST, it may still fail on others."
